```python
import xml.etree.ElementTree as et

t = et.parse('C:/Users/Student/Downloads/ELTeC-pol-master/level2/POL0001_balucki_bialy-murzyn.xml')
root = t.getroot()
wl = root.findall('.//{http://www.tei-c.org/ns/1.0}w')
```

```python
from collections import Counter
import pandas as pd

a = pd.Series([w.text for w in wl])
```

```python
b = a.value_counts().sort_values(ascending=False)
```

```python
b.index
```

```python
import matplotlib.pyplot as plt

plt.loglog(range(len(b)), b.values)
plt.title("Bialy murzyn (Bałucki, Michał) - prawo Zipfa")
plt.ylabel = "ilosc wystapien slowa"
plt.xlabel = "slowa posortowane po czestosci wystapien"
plt.show()
```
